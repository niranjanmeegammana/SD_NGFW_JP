```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set font to serif and size 14
sns.set_context("notebook", font_scale=1.4)
sns.set_style("whitegrid")
plt.rcParams["font.family"] = "serif"

# Time complexity data from the table
operations = [
    'Model Loading',
    'Data Preprocessing',
    'Model Switching',
    'Inference',
    'Logging',
    'Total Complexity'
]

# Corresponding time complexities for both scripts
ngfw_multi_model1 = [1, 2, 3, 4, 1, 5]  # O(m), O(nd), O(m), O(ndwh), O(1), O(m) + O(ndwh)
ngfw_multi_model2 = [0, 2, 0, 4, 1, 4]  # O(1), O(nd), O(1), O(ndwh), O(1), O(ndwh)

# Cumulative sums for plotting the total complexity line
cumulative_model1 = np.cumsum(ngfw_multi_model1)
cumulative_model2 = np.cumsum(ngfw_multi_model2)

# Bar width and positions
bar_width = 0.35
index = np.arange(len(operations))

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting bars for both models
bar1 = ax.bar(index - bar_width / 2, ngfw_multi_model1, bar_width, label="On-demand", color="cyan")
bar2 = ax.bar(index + bar_width / 2, ngfw_multi_model2, bar_width, label="Pre-loaded", color="pink")

# Plotting the cumulative line for both models
ax.plot(index, cumulative_model1, color='blue', marker='o', linestyle='-', label='Cumulative Total (On-demand)')
ax.plot(index, cumulative_model2, color='orange', marker='o', linestyle='-', label='Cumulative Total (Pre-loaded)')

# Adding labels, title, and grid
ax.set_xlabel("Operations", fontsize=14)
ax.set_ylabel("Time Complexity (arbitrary units)", fontsize=14)
#ax.set_title("Comparison of Time Complexities Between ngfw_multi_model1.py and ngfw_multi_model2.py", fontsize=16)
ax.set_xticks(index)
ax.set_xticklabels(operations, fontsize=12, rotation=45)
ax.legend()

# Show the plot
plt.tight_layout()
plt.savefig("NGFW Time Complexity.png", dpi=300, bbox_inches="tight")
plt.show()
```
