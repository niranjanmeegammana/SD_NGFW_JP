```python
import matplotlib.pyplot as plt
import numpy as np

# Updated frequency values
frequency = [1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]

# Interpolated CPU usage values (approximated from the trend in your image)
cpu_usage = [5, 11, 17, 22, 26, 30, 34, 38, 41, 44, 47]

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(frequency, cpu_usage, marker='o', linestyle='-', color='steelblue')
#plt.title('CPU Usage vs. Model Switching Frequency')
plt.xlabel('Switching Frequency (per minute)')
plt.ylabel('CPU Usage (%)')
plt.grid(True)

# Save the figure as a high-resolution PNG
plt.savefig('cpu_usage_vs_switching_frequency.png', dpi=300)

# Show the plot
plt.show()
```
