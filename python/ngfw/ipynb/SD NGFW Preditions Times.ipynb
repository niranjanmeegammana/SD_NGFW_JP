```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)
fruit_weights = [
    np.random.normal(130, 10, size=100),
    np.random.normal(125, 20, size=100),
    np.random.normal(120, 30, size=100),
]
labels = ['peaches', 'oranges', 'tomatoes']
colors = ['peachpuff', 'orange', 'tomato']

fig, ax = plt.subplots()
ax.set_ylabel('fruit weight (g)')

bplot = ax.boxplot(fruit_weights,
                   patch_artist=True,  # fill with color
                   tick_labels=labels)  # will be used to label x-ticks

# fill with colors
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for preloaded approach
data_preloaded = {
    'model': ['deep20', 'deep40', 'shallow20', 'shallow40'],
    'count': [103, 103, 82, 82],
    'mean': [0.090352, 0.087313, 0.086853, 0.085359],
    'std': [0.016344, 0.018301, 0.014869, 0.013267],
    'min': [0.060775, 0.051744, 0.052757, 0.059204],
    'max': [0.146490, 0.186768, 0.149457, 0.111622]
}

# Data for on-demand approach
data_on_demand = {
    'model': ['deep20', 'deep40', 'shallow20', 'shallow40'],
    'count': [87, 92, 98, 93],
    'mean': [0.093267, 0.094581, 0.080789, 0.082942],
    'std': [0.035404, 0.031635, 0.020509, 0.025319],
    'min': [0.052798, 0.049276, 0.051756, 0.052774],
    'max': [0.226229, 0.187181, 0.163403, 0.212457]
}

# Creating dataframes for both approaches
df_preloaded = pd.DataFrame(data_preloaded)
df_on_demand = pd.DataFrame(data_on_demand)

# Renaming columns to distinguish between the two approaches
df_preloaded['approach'] = 'Preloaded'
df_on_demand['approach'] = 'On-demand'

# Merging the two dataframes
df_combined = pd.concat([df_preloaded[['model', 'mean', 'min', 'max', 'approach']],
                         df_on_demand[['model', 'mean', 'min', 'max', 'approach']]],
                        ignore_index=True)

# Create a boxplot of prediction times
plt.figure(figsize=(10, 6))
plt.title('Prediction Time Distribution for Preloaded and On-demand Approaches')
box =plt.boxplot([df_preloaded['mean'], df_on_demand['mean']], labels=['Preloaded', 'On-demand'], patch_artist=True, vert=False)

# Adding custom colors

colors = ['cyan', 'peachpuff']

for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_edgecolor('black')

plt.xlabel('Prediction Time (seconds)')
plt.savefig("approch_comparison_plot.png", dpi=300, bbox_inches="tight")
plt.show()


```
